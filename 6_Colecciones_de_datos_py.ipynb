# Colecciones de Datos

Un poco m√°s a detalle sobre las collecciones de datos

```python
#Listas 
list1= [1,4,"Data",6,"college"]
list2=[]  # creates an empty list
list3=list((1,2,3))

print(list1)
print(list2)
print(list3)
```

```python
#tuplas

tuple1=(1,2,"college",9,'r','d', 7,3)
tuple2=() # creates an empty tuple
tuple3=tuple((1,8,5,9, "hello"))

print(tuple1)
print(tuple2)
print(tuple3)
```

```python
print(tuple())

print(tuple((7, 2, 3)))


print(tuple("Python"))


print(tuple([1, 2, 3]))
```

```python
#Set  / Conjuntos 
set1={1,2,3,4,5,"hello","tup",10,33, "camino"}
set2={(1,8,"python",7)}
set3={(10,8,"python",7)}

print(set1)
print(set2)
print(set3)
```

```python
#Diccionarios 
dict1={"key1":"value1","key2":"value2"}
dict2={}   # empty dictionary
dict3=dict({1:"apple",2:"cherry",3:"strawberry"})

print(dict1)
print(dict2)
print(dict3)
```

```python
# Program to demonstrate list vs set vs tuple vs dictionary
 # Initializing data structures
l=[1,2,3,4,5] # List
t=(1,2,3,4,5) # Tuple
s={1,2,3,4,5} # Set
d={"Name":"Raj","Age":25} # Dictionary

# Deletion in List
print("Value deleted from the list is",l.pop(2))
print("List is",l)

# Deletion in Tuple
print("You cannot delete an element from the tuple")
print("Tuple is",t)

# Deletion in Set
print("Value deleted from the Set is",s.pop())
print("Set is ",s)
# Deletion in Set
print("Value deleted from the Set is",s.pop())
print("Set is ",s)

# Deletion in Dictionary
print("Value deleted from the Dictionary is",d.pop("Age"))
print("Dictionary is",d)
```

# Orden de datos en colecciones

```python
# Program to demonstrate list vs set vs tuple vs dictionary
 # Initializing data structures
l=[110,2,43,54,599] # List
t=(110,2,43,54,599) # Tuple
s={110,12,43,54,599} # Set
d={"Name":"Raj","Age":25, "Location":"Mumbai"} # Dictionary

# Sorting in List
print("Unsorted list is ",l)
l.sort()
print("Sorted list is ",l)


# Sorting a Tuple
print("\nUnsorted tuple is ",t)
print("Sorted tuple is ",sorted(t))
print("Original tuple is ",t)

# Sorting a Set
print("\nUnsorted set is ",s)
print("Sorted set is ",sorted(s))
print("Original set is ",s)

# Sorting a Dictionary
print("\nUnsorted dictionary is ",d)
print("Sorted dictionary is ",sorted(d))
print("Original dictionary is ",d)
```
